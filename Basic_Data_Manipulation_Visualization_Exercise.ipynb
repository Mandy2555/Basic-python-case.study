### 50. Write a Python program to convert a string to datetime.
Sample String : Jul 1 2014 2:43PM 

Expected Output : 2014-07-01 14:43:00

```python
from datetime import datetime
Sample_String = 'Jul 1 2014 2:43PM'

datetime_object = datetime.strptime(Sample_String, "%b %d %Y %I:%M%p")
formatted_datetime = datetime_object.strftime("%Y-%m-%d %H:%M:%S")
print('Output:', formatted_datetime)
```

### 51. Write a Python program to subtract five days from current date.

Current Date : 2015-06-22

5 days before Current Date : 2015-06-17

```python
from datetime import datetime, timedelta
current_date_string = "2015-06-22"
current_date = datetime.strptime(current_date_string, "%Y-%m-%d")
five_days_before = current_date - timedelta(days=5)
print("Current Date:", current_date.date())
print("5 Days Before Current Date:", five_days_before.date())
```
